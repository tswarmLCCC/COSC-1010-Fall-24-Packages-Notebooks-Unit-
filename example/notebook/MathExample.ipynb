```python
import numpy as np
import matplotlib.pyplot as plt

# Other libraries you might need
import sympy
import scipy.integrate
```

```python
# Define the function
def f(x):
  return x**2 - 4*x + 3

# Generate data points
x = np.linspace(0, 5, 100)
y = f(x)

# Plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of a Quadratic Function')
plt.show()
```

```python
# Define the polynomial
p = sympy.Symbol('x')
polynomial = p**3 - 2*p**2 - 5

# Find the roots
roots = sympy.solve(polynomial, p)

# Print the roots
print('Roots of the polynomial:', roots)
```

```python
# Define the function
def f(x):
  return np.sin(x)

# Define the limits of integration
a = 0
b = np.pi

# Calculate the area
area = scipy.integrate.quad(f, a, b)
y = f(x)

# Plot the function

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph')
plt.show()

# Print the area
print('Area under the curve:', area[0])
```

```python
# Generate some random data points
x = np.random.uniform(0, 10, 100)
y = 2*x + 3 + np.random.randn(100)

# Fit a line to the data
coefficients = np.polyfit(x, y, 1)

# Plot the data and the fitted line
plt.plot(x, y, 'o')
plt.plot(x, np.polyval(coefficients, x), 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Line')
plt.show
```
